```python
from sklearn.model_selection import KFold
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_iris
import numpy as np

# Load dataset
X = df[['b4_SMI_24', 'b3_SMI_24', 'b2_SMI_24', 'b8_SMI_24', 'NDVI_SMI_2']].values
y = df['LandUse'].values

# Define model
model = RandomForestClassifier(random_state=42)

# Define K-Fold
k = 5
kf = KFold(n_splits=k, shuffle=True, random_state=42)

# Initialize metrics
omission_errors = []
commission_errors = []
accuracies = []

# Perform K-Fold Cross-Validation
for train_index, test_index in kf.split(X):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    
    # Train and predict
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred, labels=np.unique(y))
    TP = np.diag(cm)  # True Positives
    FN = cm.sum(axis=1) - TP  # False Negatives
    FP = cm.sum(axis=0) - TP  # False Positives
    TN = cm.sum() - (FP + FN + TP)  # True Negatives
    
    # Calculate omission and commission errors for each class
    omission = FN / (FN + TP) * 100  # Per class omission error
    commission = FP / (FP + TP) * 100  # Per class commission error
    
    # Append average omission and commission errors across all classes
    omission_errors.append(np.mean(omission))
    commission_errors.append(np.mean(commission))
    
    # Calculate accuracy
    accuracy = accuracy_score(y_test, y_pred)
    accuracies.append(accuracy)

# Calculate mean and standard error
mean_omission_error = np.mean(omission_errors)
mean_commission_error = np.mean(commission_errors)
mean_accuracy = np.mean(accuracies)

std_error_accuracy = np.std(accuracies) / np.sqrt(k)

# Print results
print(f"Mean Omission Error: {mean_omission_error:.2f}%")
print(f"Mean Commission Error: {mean_commission_error:.2f}%")
print(f"Mean Accuracy: {mean_accuracy:.2f}")
print(f"Standard Error of Accuracy: {std_error_accuracy:.4f}")
```
