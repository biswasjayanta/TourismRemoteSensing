```python
model_desc = '''
# Measurement model
MAWEI =~ WAI + SPM
NDWI =~ WAI + SPM
SPM =~ WAI

# Structural model
WAI ~ NDWI + MAWEI + SPM
MAWEI ~ NDWI + SPM
SPM ~ NDWI + MAWEI
'''
```

```python
# Define the SEM model
sem_model = """
    # Auto-regressive paths
    WAI_20 ~ WAI_18
    SPM_20 ~ SPM_18
    MAWEI_20 ~ MAWEI_18
    NDWI_20 ~ NDWI_18

    WAI_22 ~ WAI_20
    SPM_22 ~ SPM_20
    MAWEI_22 ~ MAWEI_20
    NDWI_22 ~ NDWI_20

    WAI_24 ~ WAI_22
    SPM_24 ~ SPM_22
    MAWEI_24 ~ MAWEI_22
    NDWI_24 ~ NDWI_22

    # Cross-lagged paths
    SPM_20 ~ WAI_18
    MAWEI_20 ~ WAI_18 + SPM_18
    NDWI_20 ~ WAI_18 + SPM_18 + MAWEI_18

    SPM_22 ~ WAI_20
    MAWEI_22 ~ WAI_20 + SPM_20
    NDWI_22 ~ WAI_20 + SPM_20 + MAWEI_20

    SPM_24 ~ WAI_22
    MAWEI_24 ~ WAI_22 + SPM_22
    NDWI_24 ~ WAI_22 + SPM_22 + MAWEI_22

    # Within-time-period relationships
    SPM_18 ~ WAI_18
    MAWEI_18 ~ WAI_18 + SPM_18
    NDWI_18 ~ WAI_18 + SPM_18 + MAWEI_18

    SPM_20 ~ WAI_20
    MAWEI_20 ~ WAI_20 + SPM_20
    NDWI_20 ~ WAI_20 + SPM_20 + MAWEI_20

    SPM_22 ~ WAI_22
    MAWEI_22 ~ WAI_22 + SPM_22
    NDWI_22 ~ WAI_22 + SPM_22 + MAWEI_22

    SPM_24 ~ WAI_24
    MAWEI_24 ~ WAI_24 + SPM_24
    NDWI_24 ~ WAI_24 + SPM_24 + MAWEI_24
"""


# Fit the model
model = Model(sem_model)
res = model.fit(data)

# Display results
print(model.inspect())
```

```python
# Analyze parameter estimates
ins = model.inspect()
print(ins)
```

```python
# Calculate additional SEM statistics
stats = calc_stats(model)  # Use calc_stats directly
print(stats)
```

```python
### Visualize Structural Equation Model
# Plot the SEM model with different colors
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Scatter plot with different colors for different attributes
scatter = ax.scatter(
    df['NDWI'],
    df['MAWEI'],
    df['WAI'],
    c=df['WAI'],  # Use Test_Scores as a color indicator
    cmap='viridis',  # Choose a color map
    label='WAI'
)

ax.set_xlabel('NDWI')
ax.set_ylabel('MAWEI')
ax.set_zlabel('WAI')

# Add colorbar
cbar = plt.colorbar(scatter, ax=ax, label='WAI')
cbar.set_label('WAI')

plt.show()
```
